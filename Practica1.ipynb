```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.optimize import curve_fit
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/SantiagoJulioD/Personal/refs/heads/main/datosp1.csv')
```

```python
V1 = df.V1.values
V2 = df.V2.values
```

```python
df
```

```python
Vd = V2*3./4095
Id = (V1 - V2)*3./4095*1000
```

```python
func = lambda x,a,b,c: a*np.exp(b*x)+c
```

```python
params,_ = curve_fit(func,Vd,Id)
```

```python
plt.plot(Vd,Id)
plt.plot(Vd,func(Vd,*params))
```
