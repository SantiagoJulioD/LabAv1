```python
import numpy as np 
import pandas as pd 
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/SantiagoJulioD/LabAv1/refs/heads/main/LabAv1-P1/datosp1.csv')
df['V1_2'] = df.apply(lambda x:np.random.poisson(x['V1'],1)[0],axis=1)
df['V2_2'] = df.apply(lambda x:np.random.poisson(x['V2'],1)[0],axis=1)
df
```

```python
import matplotlib.pyplot as plt
plt.plot(df['V1_2'],df['V2_2'])
plt.plot(df['V1'],df['V2'])
```

```python
V1 = df.V1.values
V2 = df.V2.values
V1_2 = df.V1_2.values
V2_2 = df.V2_2.values
```

```python
Vd = V2*3./4095
Id = (V1 - V2)*3./4095*1000
Vd_2 = V2_2*3./4095
Id_2 = (V1_2 - V2_2)*3./4095*1000
```

```python
df_2 = pd.DataFrame({'Vd_1':Vd,
                     'Vd_2':Vd_2,
                     'Id_1':Id,
                     'Id_2':Id_2})

df_2.to_csv('datap1.csv')
```

```python
plt.plot(df_2.Vd_1,df_2.Id_1)
plt.plot(df_2.Vd_2,df_2.Id_2)
```
