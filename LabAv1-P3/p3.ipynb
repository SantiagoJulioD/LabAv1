```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
df = pd.read_csv('/home/santiago/Documentos/LabAv1/practica3.csv')
df['Ángulo'] = df['Ángulo']*np.pi/180
df['Luz'] = df['Luz']/max(df['Luz'])
```

```python
def func(theta,A,B,C,D):
    return A*(np.cos(B*(theta-C)))**2+D
```

```python
p0 = (1.,1.,np.pi/4,0.)
params,_ = curve_fit(func,df['Ángulo'],df['Luz'],)
params
```

```python
x = np.linspace(0,np.pi,100)
y = func(x,*params)
```

```python
plt.plot(df['Ángulo'],df['Luz'],'r.')
plt.plot(x,y)
```
